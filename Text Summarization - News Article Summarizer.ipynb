```python
# Use Case: Automatic News/Article Summarization
# Business Impact: Save 80% reading time, extract key insights

from transformers import pipeline
import gradio as gr

# Initialize summarization pipeline (free)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", framework="pt")

def summarize_text(text, max_length=130, min_length=30):
    if not text.strip():
        return "Please enter text to summarize!"
    
    if len(text.split()) < 50:
        return "⚠️ Text too short. Please provide at least 50 words for meaningful summarization."
    
    try:
        # Generate summary
        summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
        
        original_words = len(text.split())
        summary_words = len(summary[0]['summary_text'].split())
        compression_ratio = ((original_words - summary_words) / original_words) * 100
        
        return f"""
## 📄 AI-Generated Summary

{summary[0]['summary_text']}

---
**📊 Compression Stats:**
- Original: {original_words} words
- Summary: {summary_words} words  
- Space saved: {compression_ratio:.1f}%
        """
    except Exception as e:
        return f"❌ Error: {str(e)}"

# Create interface
interface = gr.Interface(
    fn=summarize_text,
    inputs=[
        gr.Textbox(
            placeholder="Paste your article, research paper, or long text here...",
            label="Text to Summarize",
            lines=10
        ),
        gr.Slider(50, 200, value=130, label="Max Summary Length"),
        gr.Slider(20, 80, value=30, label="Min Summary Length")
    ],
    outputs=gr.Markdown(),
    title="🤖 AI Text Summarizer",
    description="Summarize articles, papers, reports instantly using BART AI model",
    examples=[
        ["""Artificial intelligence (AI) has revolutionized numerous industries, from healthcare to finance, transportation, and entertainment. In healthcare, AI algorithms can analyze medical images with unprecedented accuracy, helping doctors diagnose diseases earlier and more precisely. Machine learning models can predict patient outcomes, optimize treatment plans, and even discover new drugs. In the financial sector, AI powers algorithmic trading, fraud detection systems, and personalized banking services. Transportation has been transformed by autonomous vehicles, which use computer vision and sensor fusion to navigate safely. The entertainment industry leverages AI for content recommendation systems, video game AI, and even AI-generated music and art. However, the rapid advancement of AI also raises important ethical questions about job displacement, privacy, bias in algorithms, and the need for proper regulation. As AI continues to evolve, society must balance innovation with responsible development to ensure these powerful technologies benefit everyone."""]
    ]
)

if __name__ == "__main__":
    interface.launch(share=True)
```
